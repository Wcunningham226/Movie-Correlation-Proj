```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np
import re

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the config of the plots we will create


# Read in the data

df = pd.read_csv(r'/Users/willcunningham/Desktop/google_data_analytics_cert/Portfolio_project/PortfolioProject_4/movies.csv')
```

```python
# Let's look at data

df.head()
```

```python
# Look for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change data type of columns / Capital I in int64 allows change but also keeps NaN

df['budget'] = df['budget'].astype('Int64')

df['gross'] = df['gross'].astype('Int64')
```

```python
# Confirm type changes

df.head()
```

```python
# Preprocess a new column to discard non relevant data i.e. country in released column

df['yearcorrect'] = pd.to_datetime(df['released'].str.extract(r'(\w+ \d+, \d+)', expand=False), format='%B %d, %Y')

df.head()
```

```python
# Create correct year column

df['yearcorrect'] = df['yearcorrect'].astype(str).str[:4]

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot with budget vs gross 

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot budget using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized 
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
